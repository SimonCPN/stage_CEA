```python
data = np.load(filedir+filename,allow_pickle=True)
print(data.files)

Lambda_Train = data['Lambda_Train']
Lambda_Test = data['Lambda_Test']
param_train = data['param_train'][()]
param_test = data['param_test'][()]

print(Lambda_Train.shape)
print(param_train.keys())
```

# Display parameters and barycentric coordinates

```python
i_wf = 0 # number of the waveform from the training set you want to print the parameters of

print('f0 (Hz):',param_test['f0'][i_wf])
print('EclLat (rad):',param_test['EclLat'][i_wf])
print('EclLon (rad):',param_test['EclLon'][i_wf])
print('iota (rad):',param_test['iota'][i_wf])
print('psi (rad):',param_test['psi'][i_wf])
print('phi0 (rad):',param_test['phi0'][i_wf])
```

```python
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(Lambda_Train[i_wf,:,0])
plt.grid()
plt.title(r'Re(A)')
plt.subplot(222)
plt.plot(Lambda_Train[i_wf,:,1])
plt.grid()
plt.title(r'Im(A)')
plt.subplot(223)
plt.plot(Lambda_Train[i_wf,:,2])
plt.grid()
plt.title(r'Re(E)')
plt.subplot(224)
plt.plot(Lambda_Train[i_wf,:,3])
plt.grid()
plt.title(r'Im(E)')
plt.show()
print(len(Lambda_Train[i_wf,:,0]))
```
